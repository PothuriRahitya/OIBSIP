```python
#importing required python libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
columns=['id','sepal length','sepal width','petal length','petal width','Class_labels']
#loading dataframe by reading the dataset in csv file
df=pd.read_csv('dataset_iris.csv',names=columns)
df.head()
```

```python
df.shape
#to know count of rows and columns of given dataset
```

```python
df.describe()
#data visualization
```

```python
sns.pairplot(df, hue='Class_labels')
```

```python
#seperating the values of sample and class
data=df.values
x=data[:,1:5]
y=data[:,5]
```

```python
#allocating dataset values for testing and training
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#we can use logistic regression to get accurate results for classification purposes
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
#accuracy of this model
prediction=model.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction)*100)
```

```python
#printing test class and predicted class by the machine learning model
for i in range(len(prediction)):
  print(y_test[i],prediction[i])
```

```python
#testing the model with another value
x_new=np.array([[3, 2, 1, 0.2], [  4.9, 2.2, 3.8, 1.1 ], [  5.3, 2.5, 4.6, 1.9 ]])
prediction1=model.predict(x_new)
print(format(prediction1))
```

```python
#report
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction))
```
